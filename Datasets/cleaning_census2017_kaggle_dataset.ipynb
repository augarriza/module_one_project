```python
import pandas as pd
```

```python
census = pd.read_csv("acs2017_census_tract_data.csv")
#This dataset contains data about Puerto Rico, but we´ll drop these rows since our other dataframes don´t
census.drop(census[census["State"] == "Puerto Rico"].index, inplace=True)
census.columns
```

```python
# Employed: Number of employed (16+)

df_men_women = census[["State", "TotalPop", "Men", "Women", "Employed"]]
df_pop_gender = (df_men_women
                 .groupby("State")
                 .agg({"TotalPop":"sum",
                       "Men":"sum",
                       "Women":"sum",
                       "Employed":"sum"})
                 .reset_index())

df_pop_gender.head()
```

```python
df_etnic = census[['State', 'Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']]
df_etnic = (df_etnic
            .groupby('State')
            .agg({'Hispanic':'mean',
                  'White':'mean',
                  'Black':'mean',
                  'Native':'mean',
                  'Asian':'mean',
                  'Pacific':'mean'})
            .reset_index())

df_etnic.head()
```

```python
# Poverty: % under poverty level
# Unemployment: unemployment rate (%)

df_income = census[['State', 'IncomePerCap', 'Poverty', 'Unemployment']]
df_income = (df_income
             .groupby('State')
             .agg({'IncomePerCap':'mean',
                   'Poverty':'mean',
                   'Unemployment':'mean'})
             .reset_index())

df_income.head()
```

```python
# PrivateWork: % employed in private industry
# PublicWork: % employed in public jobs
# FamilyWork: % in unpaid family work

df_employment = census[['State','PrivateWork','PublicWork','SelfEmployed','FamilyWork']]
df_employment = (df_employment
             .groupby('State')
             .agg({'PrivateWork':'mean',
                   'PublicWork':'mean',
                   'SelfEmployed':'mean',
                   'FamilyWork':'mean'})
             .reset_index())

df_employment.head()
```

```python
# Professional: % employed in management, business, science, and arts
# Service: % employed in service jobs
# Office: % employed in sales and office jobs
# Construction: % employed in natural resources, construction, and maintenance
# Production: % employed in production, transportation, and material movement

df_sector = census[['State','Professional','Service','Office','Construction','Production']]
df_sector = (df_sector
             .groupby('State')
             .agg({'Professional':'mean',
                   'Service':'mean',
                   'Office':'mean',
                   'Construction':'mean',
                   'Production':'mean'})
             .reset_index())

df_sector.head()
```
