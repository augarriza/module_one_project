```python
import pandas as pd
```

```python
df = pd.read_pickle("../../Pickles/nytimes_cdcapi_combined.pkl")
df.head()
```

We want to compute the mortality rate of covid

```python
df_mrt = df.assign(nyt_mortality=0, cdc_mortality=0)
ny_deaths = df_mrt["nyt_covid_deaths"].tolist()
cdc_deaths = df_mrt["cdc_covid_deaths"].tolist()
cases = df_mrt["covid_cases"].tolist()

df_mrt["nyt_mortality"]=[float((d/c)*100) for d,c in zip(ny_deaths,cases)]
df_mrt["cdc_mortality"]=[float((d/c)*100) for d,c in zip(cdc_deaths,cases)]

    

df_mrt.head()
```

```python
df_mrt.to_pickle("../../Pickles/nytimes_cdcapi_withmortality.pkl")
df=pd.read_pickle("../../Pickles/nytimes_cdcapi_withmortality.pkl")
df.head()
```
