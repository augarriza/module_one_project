```python
import pandas as pd
```

```python
df_pop = pd.read_pickle('../../Pickles/df_census_tp_gender.csv')
df_pop.head()
```

```python
df_inc = pd.read_pickle('../../Pickles/df_census_income.csv')
df_inc.head()
```

```python
df_age = pd.read_pickle('../../Pickles/df_census_groupage.csv')
df_age.head()
```
