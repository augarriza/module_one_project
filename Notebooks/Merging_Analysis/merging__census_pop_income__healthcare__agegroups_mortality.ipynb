```python
import pandas as pd
```

```python
df_pop = pd.read_pickle('../../Pickles/df_census_tp_gender.csv')
```

```python
df_inc = pd.read_pickle('../../Pickles/df_census_income.csv')
```

```python
df_age = pd.read_pickle('../../Pickles/df_census_groupage.csv')
```

```python
df_hlt = pd.read_pickle('../../Pickles/healthcare_percentage.pkl')
```

```python
df_mor = pd.read_pickle('../../Pickles/cdc_morality_to_merge.pkl')
df_mor = (df_mor.
          drop(columns='Year')
          .rename(columns={'Deaths':'Deaths Feb-Mar ´20',
                                                     'Death Difference':'Difference Feb-Mar ´18'}))
```

```python
final_df = (df_pop.merge(df_inc, how='left', on='State')
            .merge(df_hlt, how='left', on='State')
            .merge(df_age, how='left', on='State')
            .merge(df_mor, how='left', on='State')
            .reset_index(drop=True))

final_df.head()
```

```python
final_df.to_pickle('../../Pickles/final_df.pkl')
```
