```python
import pandas as pd
import numpy as np
```

# Starting to analyze and process both sets of mortality data

## All Deaths with Covid

```python
df = pd.read_pickle("Datasets/cdc_mortality_df.pkl")
```

```python
df.head()
```

```python
df = df.astype({"Deaths":"int64"})
```

```python
df["Death Difference"] = df["Deaths"].diff()
```

```python
df["Death Difference"].iloc[::7] = np.NaN
```

```python
df.loc[(df["Death Difference"]>0) & (df["Year"]==2020)]
```

```python
df.loc[df["State"] == "Florida"]
```

## All Deaths without Covid

```python
df2 = pd.read_pickle("Datasets/cdc_mortality_no_covid_df.pkl")
```

```python
df2["Death Difference no Covid"] = df2["Deaths"].diff()
```

```python
df2["Death Difference no Covid"].iloc[::7] = np.NaN
```

```python
df2.loc[(df2["Death Difference no Covid"]>0) & (df2["Year"]==2020)]
```

```python
df.merge(df2)
```
